```python
# === Task 1: Load and Explore the Dataset ===

# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
# Step 2: Load the iris dataset
iris = load_iris()

# Step 3: Convert to pandas DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Step 4: Inspect the dataset
print("First 5 rows:")
display(df.head())

print("\nDataset Info:")
print(df.info())

print("\nMissing Values Check:")
print(df.isnull().sum())
```

```python
# === Task 2: Basic Data Analysis ===

print("\nDescriptive Statistics:")
display(df.describe())

print("\nAverage measurements grouped by species:")
display(df.groupby("species").mean())
```

```python
# === Task 3: Visualizations ===

# 1. Line Chart (Sepal length across samples)
plt.figure(figsize=(8,4))
plt.plot(df['sepal length (cm)'])
plt.title("Sepal Length across Samples")
plt.xlabel("Sample Index")
plt.ylabel("Sepal Length (cm)")
plt.show()
print("Observation: Sepal length values generally range between 4.5 cm and 8 cm. "
      "There is noticeable variation across samples, but no obvious outliers.")

```

```python
# === Task 3: Visualizations ===

# 2. Bar Chart (Average petal length per species)
plt.figure(figsize=(6,4))
df.groupby("species")["petal length (cm)"].mean().plot(kind="bar", color=['skyblue','orange','green'])
plt.title("Average Petal Length per Species")
plt.xlabel("Species")
plt.ylabel("Average Petal Length (cm)")
plt.show()
print("Observation: Setosa has the smallest average petal length, "
      "while Virginica has the largest. Versicolor falls in between.")

```

```python
# === Task 3: Visualizations ===

# 3. Histogram (Distribution of sepal width)
plt.figure(figsize=(6,4))
plt.hist(df['sepal width (cm)'], bins=15, color='purple', alpha=0.7)
plt.title("Distribution of Sepal Width")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")
plt.show()
print("Observation: Sepal width is mostly concentrated between 2.5 cm and 3.5 cm, "
      "with fewer flowers having widths below 2.5 cm or above 4.0 cm.")
```

```python
# === Task 3: Visualizations ===

# 4. Scatter Plot (Sepal Length vs Petal Length)
plt.figure(figsize=(6,4))
sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', hue='species', data=df, palette='Set1')
plt.title("Sepal Length vs Petal Length by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend()
plt.show()
print("Observation: Setosa species are clearly separated from the other two "
      "due to their shorter petal lengths. Versicolor and Virginica overlap somewhat, "
      "but Virginica tends to have longer petals.")
```

```python
# === Final Summary & Conclusion ===

print("\n===== Summary & Conclusion =====\n")
print("1. The Iris dataset consists of 150 samples across 3 species: Setosa, Versicolor, and Virginica.")
print("2. Descriptive statistics showed that Virginica generally has the largest sepal and petal sizes, while Setosa has the smallest.")
print("3. Grouped analysis confirmed that average petal length and width are distinguishing features across species.")
print("4. Visualizations highlighted clear separation of Setosa from the other species, particularly in petal dimensions.")
print("5. While Versicolor and Virginica overlap in sepal length, Virginica consistently exhibits longer petals.")
print("\nOverall, the Iris dataset demonstrates how basic data analysis and visualization can reveal clear patterns, "
      "supporting its continued use as a benchmark dataset in machine learning and classification tasks.")
```
